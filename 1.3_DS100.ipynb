```python
import matplotlib.pyplot as plt
import numpy as np

fertility=[3.151,1.443,1.261,1.82,2.434,2.001]
gdp=[5641,12822,72056,4486,8498,20398]
population=[93.2,69.1,50.9,87.8,239.9,48.0]
color=['olive','pink','chartreuse','plum','wheat','turquoise']

np_pop=np.array(population)
np_pop=np_pop*10

plt.scatter(gdp,fertility,c=color,s=np_pop)
plt.xscale('log')
plt.xlabel('GDP Per Capita')
plt.ylabel('Fertility')
plt.title('Fertility as a function of GDP per Capita for South East Asia countries')
tick_val=[1000,10000,100000]
tick_lab=['1k','10k','100k']
plt.xticks(tick_val,tick_lab)
plt.show()
```
